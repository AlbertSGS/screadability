```python
import pandas as pd
```

```python
xl_file = pd.ExcelFile("dataset/ground_truth_label.xlsx")
```

```python
dfs = [{"df": pd.read_excel(xl_file, index_col=None, sheet_name=sheet_name, usecols="A, B, C"), "type": sheet_name} for sheet_name in xl_file.sheet_names]
```

```python
for i in range(len(dfs)):
    dfs[i]["df"].insert(0, "type", dfs[i]["type"])
    dfs[i] = dfs[i]["df"]
```

```python
full_df = pd.concat(dfs, ignore_index=True)
```

```python
dfs_true = full_df[full_df["ground truth"] == 1].copy()
```

```python
dfs_true.sort_values(by=["type", "file"], inplace=True)
```
